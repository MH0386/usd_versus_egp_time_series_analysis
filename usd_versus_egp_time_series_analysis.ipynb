```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
try:
    data = pd.read_csv('USD_EGP Historical Data.csv')
except:
    data = pd.read_csv('/kaggle/input/usd-to-egp/USD_EGP Historical Data1.csv')
data
```

```python
data.info()
```

```python
data['Change %'] = data['Change %'].astype(str)
data['Change %'] = data['Change %'].str.replace('%', '').astype(float)
data
```

```python
data.info()
```

```python
plt.figure(figsize=(20, 6))
plt.plot(data.Date, data.Price)
plt.plot(data.Date, data.High)
plt.plot(data.Date, data.Low)
plt.title('USD/EGP')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(['Price', 'High', 'Low', 'Change %'])
plt.show()
```

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=data.Date, y=data.Price,
        mode='lines',
        name='Price'
    )
)
fig.add_trace(
    go.Scatter(
        x=data.Date, y=data.High,
        mode='lines',
        name='High'
    )
)
fig.add_trace(
    go.Scatter(
        x=data.Date, y=data.Low,
        mode='lines', 
        name='Low'
    )
)
fig.show()
```
